```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
dias = np.arange(1, 31)
vendas = np.random.randint(50, 150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia': dias, 'Vendas': vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_vendas['Dia'], df_vendas['Vendas'], marker='o', linestyle='-', color='b')

plt.title('Acompanhamento de Vendas Diárias', fontsize=16)
plt.xlabel('Dia do Mês', fontsize=14)
plt.ylabel('Vendas Diárias', fontsize=14)

plt.grid(True)
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Móveis', 'Livros']
vendas_totais = [1200, 850, 950, 500, 700]

df_vendas_categorias = pd.DataFrame({'Categoria': categorias, 'Vendas': vendas_totais})
```

```python
df_vendas_categorias
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_vendas_categorias['Categoria'], df_vendas_categorias['Vendas'], color='skyblue')

plt.title('Vendas Totais por Categoria de Produto - Mês Atual', fontsize=16)
plt.xlabel('Categoria de Produto', fontsize=14)
plt.ylabel('Vendas Totais', fontsize=14)
```

```python
marcas = ['Apple', 'Samsung', 'Xiaomi', 'OnePlus', 'Outras']
participacao_mercado = [35, 30, 15, 10, 10]

df_participacao = pd.DataFrame({'Marca': marcas, 'Participacao': participacao_mercado})
```

```python
df_participacao
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df_participacao['Participacao'], labels=df_participacao['Marca'], autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightgreen', 'orange', 'lightcoral'])

plt.title('Participacao de Mercado de Smartphones', fontsize=16)

plt.show()
```
